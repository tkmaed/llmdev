```python
# 掛け算を行う関数
def multiply(a, b):
    return a * b

# 2乗を行う関数
def square(number):
    return multiply(number, number)

print(square(5))
```

```python
def introduce(**kwargs):
    for key, value in kwargs.items():
        print(f"{key} : {value}")

introduce(name="太郎", age=20, hobby="読書")
```

```python
def add(*args):
    result = sum(args)
    print(f"合計:{result}")

add(1, 2, 3)
add(5, 10, 15, 20)
```

```python
def is_palindrome(s):
    if s == s[::-1]:
        print(f"「{s}」は回文です！")
    else:
        print(f"「{s}」は回文ではないです。")

is_palindrome("にんてんどううどんてんに")

is_palindrome("Pythonは便利だが遅い")

is_palindrome("no devil lived on")
```
