英単語の入力を受け付けます。未入力でリターン（エンター）キーを押されるまで、英単語の受付を繰り返します。受付が完了したら、入力された英単語をABC順に並べ替えて表示します。また、すべての英単語に含まれるアルファベットごとの数も表示します。

```python
input_value = ""
array = []
dict = {'a':0, 'b':0, 'c':0, 'd':0, 'e':0, 'f':0, 'g':0, 'h':0, 'i':0, 'j':0, 'k':0, 'l':0, 'm':0, 'n':0, 'o':0, 'p':0, 'q':0, 'r':0, 's':0, 't':0, 'u':0, 'v':0, 'w':0, 'x':0, 'y':0, 'z':0}

while True:
    input_value = input("英単語を入力してください: ")
    if input_value == "":
        break
    for v in input_value:
        dict[v] += 1
    array.append(input_value)

array.sort()

print(f"入力した英単語: {array}")

for alphabet in dict.keys():
    print(f"{alphabet}が{dict[alphabet]}個ありました")
```
