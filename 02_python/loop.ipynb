```python
i = 0
while True:
    i += 1
    print(i)
    if i >= 10:  # 条件を満たしたらループを終了
        break
```

```python
i = 0

while True:
    i += 1

    if i > 10:
        break

    if i % 2 == 1:
        continue

    print(i)

print("処理を終了します")
```

```python
i = 0

while i < 10:
    i += 1
    print(i)

print("処理を終了します")
```
