```python
try:
    num = int(input("整数を入力してください: "))
    result = 100 / num
except ZeroDivisionError:
    print("エラー: 0 で割ることはできません")
except ValueError:
    print("エラー: 整数を入力してください")
else:
    print(f"計算結果: {result}")
finally:
    print("プログラム終了")
```

```python
a = 100
b = 0

try:
    # 割り算を行う前にチェック
    if b == 0:
        raise ZeroDivisionError("0 で割り算しようとしています")
    
    # 安全な場合のみ割り算を行う
    print(a / b)
except ZeroDivisionError as e:
    print(e)
```

```python
try:
    num = int(input("整数を入力してください: "))
    result = 100 / num
    print(f"100 を {num} で割った結果は {result} です")
except ZeroDivisionError:
    print("エラー: 0 で割ることはできません")
except ValueError:
    print("エラー: 整数を入力してください")
except Exception as e:
    # 予期しない例外をキャッチ
    print(f"予期しないエラーが発生しました: {e}")
```

```python
try:
    # 例外が発生するかもしれない処理
    num = int(input("整数を入力してください: "))
    result = 100 / num
    print(f"100 を {num} で割った結果は {result} です")
except ZeroDivisionError:
    # 例外が発生した場合の処理
    print("エラー: 0 で割ることはできません")
except ValueError:
    print("エラー: 整数を入力してください")
```
